# Fraudulent Claim Detection - Corrected Solution Notebook

## 1. Data Preparation

### 1.0 Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
```

### 1.1 Load Dataset

```python
df = pd.read_csv('insurance_claims.csv')
df.head()
```

```python
# Dataset shape
df.shape
```

```python
# Column information
df.info()
```

```python
# Basic statistics
df.describe()
```

## 2. Data Cleaning

### 2.1 Handle Null Values

```python
# Replace '?' with NaN
df = df.replace('?', np.nan)
```

```python
# Check null values
df.isnull().sum()
```

```python
# Forward fill missing values
df = df.ffill()
```

### 2.2 Handle Redundant Columns

```python
# Drop unknown column if present
df.drop('_c39', axis=1, inplace=True)
```

```python
# Drop identifier column
df.drop('policy_number', axis=1, inplace=True)
```

### 2.3 Fix Data Types

```python
# Convert date columns
df['policy_bind_date'] = pd.to_datetime(df['policy_bind_date'])
df['incident_date'] = pd.to_datetime(df['incident_date'])
```

```python
# Convert fraud label to binary
df['fraud_reported'] = df['fraud_reported'].map({'Y':1,'N':0})
```

## 3. Train Validation Split

```python
# Define features and target
X = df.drop('fraud_reported', axis=1)
y = df['fraud_reported']
```

```python
# Drop datetime columns before modelling
X = X.drop(['policy_bind_date','incident_date'], axis=1)
```

```python
# Convert categorical variables to dummy variables
X = pd.get_dummies(X, drop_first=True)
```

```python
# Train validation split
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

## 5. Feature Engineering

```python
# Fill remaining missing values
X_train = X_train.fillna(X_train.mean())
X_val = X_val.fillna(X_train.mean())

# Create scaled copies
X_train_scaled = X_train.copy()
X_val_scaled = X_val.copy()

numeric_cols = X_train.select_dtypes(include=['int64','float64']).columns

scaler = StandardScaler()

X_train_scaled[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_val_scaled[numeric_cols] = scaler.transform(X_val[numeric_cols])
```

## 6. Model Building

### Logistic Regression

```python
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train_scaled, y_train)
```

```python
log_pred = log_model.predict(X_val_scaled)

print(confusion_matrix(y_val, log_pred))
print(classification_report(y_val, log_pred, zero_division=0))
```

### ROC Curve

```python
log_prob = log_model.predict_proba(X_val_scaled)[:,1]

fpr, tpr, thresholds = roc_curve(y_val, log_prob)

plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()

print('ROC AUC Score:', roc_auc_score(y_val, log_prob))
```

### Random Forest

```python
rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'
)

rf_model.fit(X_train, y_train)
```

```python
rf_pred = rf_model.predict(X_val)

print(confusion_matrix(y_val, rf_pred))
print(classification_report(y_val, rf_pred, zero_division=0))
```

### Random Forest ROC AUC

```python
rf_prob = rf_model.predict_proba(X_val)[:,1]

print('Random Forest ROC AUC:', roc_auc_score(y_val, rf_prob))
```
